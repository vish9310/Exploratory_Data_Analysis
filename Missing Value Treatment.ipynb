```python
from missingpy import KNNImputer
```

```python
st4={'X1':pd.Series([1,2,3]),'X2':pd.Series([3,4,3,3]),'X3':pd.Series(['NaN',6,5,3]),'X4':pd.Series([8,7,8,3])}
type(st4)
```

```python
df4=pd.DataFrame(st4)
df4
```

```python
#creating the inmputation object

imputer = KNNImputer(missing_values='NaN', metric='masked_euclidean', n_neighbors=3, weights='distance')

new_df4 = imputer.fit_transform(df4)


new_df4
```
