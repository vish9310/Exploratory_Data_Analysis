```python
import pandas as pd
import os
path="C:/Users/linso/programs"

os.chdir(path)
os.listdir(path)
homicide=pd.read_csv("homicide.csv")
```

# Q1.number of incidents per state per year.

```python
state_wise=homicide.groupby(['State','Year'])['Record ID'].count().reset_index()
state_wise=state_wise.rename(columns={"Record ID":"Incident_Count"})
state_wise.sort_values(by="Incident_Count",ascending= False)
```

# Q2.which month wife's have more tendency to hurt husband.

```python
monthwife=homicide[["Month","Record ID","Victim Sex","Relationship",'Perpetrator Sex']]
monthwife=monthwife[(monthwife['Victim Sex']=="Male") & (monthwife['Relationship']=='Husband') & (monthwife['Perpetrator Sex']=='Female')]
monthwife=monthwife.groupby(['Month'])["Record ID"].count().reset_index()
monthwife
```

# 3.number of incident involving exhusband vs ex wife[older or younger]

```python
ex=homicide[["Month","Record ID","Victim Sex","Relationship",'Perpetrator Sex']]
ex=ex[(ex['Victim Sex']=="Male") & (ex['Relationship']=='Ex-Husband') & (ex['Perpetrator Sex']=='Female')
      | (ex['Victim Sex']=="Female") & (ex['Relationship']=='Ex-Wife') & (ex['Perpetrator Sex']=='Male')]
ex=ex.groupby(["Relationship"])["Record ID"].count().reset_index()  #to know the split
ex
ex["Record ID"].sum()   # to know the total count of incidents
```

# 4.number of incident between empliyee and employer.

```python
emp=homicide[["Month","Record ID","Relationship"]]
emp=emp[(emp["Relationship"]=="Employee")|(emp["Relationship"]=="Employer")]
emp=emp.groupby(["Relationship"])["Record ID"].count().reset_index()
emp  # detail 
emp['Record ID'].sum()   # total number of incident between employee and employer
```

# 5.number of incident based on family relationship(ex=wife killed by husband)

```python
family = homicide[['Record ID','Relationship','Perpetrator Sex']]
family = family[((family['Relationship']=='Brother') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Brother') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Sister') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Sister') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Mother') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Mother') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Son') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Son') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Daughter') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Daughter') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Wife') & (family['Perpetrator Sex']=='Male'))|
                   ((family['Relationship']=='Husband') & (family['Perpetrator Sex']=='Female'))|
                   ((family['Relationship']=='Father') & (family['Perpetrator Sex']=='Female'))]
family = family.groupby(['Relationship'])['Record ID'].count().reset_index()
family  #report of incident gruped by different family members.


family['Record ID'].sum()   #total incidents
```
