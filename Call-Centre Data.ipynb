```python
import os
import pandas as pd

path= r"C:\Users\Lenovo\Desktop\Call Centre Analysis- Data set"
os.chdir(path)
os.listdir(path)

Calling_data=pd.read_excel("Calling Data.xlsx")
Enr_data=pd.read_excel("Enrollment Data.xlsx")
Id_table = pd.read_excel("ID Table.xlsx")
Leads = pd.read_excel('Leads.xlsx')
Course_name =pd.read_excel("Course Name.xlsx")
```

```python
#number of leads for each source

sourse_wise_leads=Leads.groupby(['Lead Source ID']).size().reset_index()
#Changin the columns name
sourse_wise_leads=sourse_wise_leads.rename(columns = {0: 'Total Leads'})
sourse_wise_leads.head()
```

```python
#Getting the source id for each lead in enrollment table;
enrollment=pd.merge(Enr_data,Leads[['Lead ID','Lead Source ID']],left_on="Lead ID", right_on="Lead ID",how="inner")

enrollment
```

```python
source_wise_enrollment=enrollment.groupby(['Lead Source ID']).size().reset_index()
source_wise_enrollment.head()
```

```python
final_table=pd.merge(sourse_wise_leads,source_wise_enrollment,left_on='Lead Source ID',right_on='Lead Source ID',how='left')
final_table
Final_table=final_table.rename(columns = {0:'Total Enroll'})
Final_table.head()
```

```python
Final_table['Convertion Ratio']=Final_table['Total Enroll'] / Final_table['Total Leads']
Final_table.sort_values(by='Convertion Ratio',ascending=False).head()
```
